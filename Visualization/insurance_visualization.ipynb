### Insurance Dataset: Exploratory Data Analysis
In this notebook, we are going to analyse the insurance dataset from kaggle. As of now we are going to do only exploratory analysis.

We will try to answer the following questions:
- Who are paying highest premium charges?
- Which of the parameter contributes most in higher premium charges? example smoking, gender, region etc
- Does bmi affects insurance charges differently for male and female?
- Does premium charges increases with age?

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
os.chdir('C://Users//panka//github_repo//datasets//kaggle_dataset/')
os.getcwd()

# the dataset is available through kaggle
df = pd.read_csv('insurance.csv')
df.head()
```

Let's plot the histogram and check the distribution of the dataset

```python
sns.displot(data=df, x='charges', kde=True)
sns.displot(df['charges'], kind='kde', rug=True)
```

#### Who are paying the highest premium?
Let's look at the histogram of the data with segmentation of smoker and non-smoker. As evident from the histogram, almost all the higer premium are paid by smokers.

```python
# sns distplot has been deprecated, using displot
# plots can be drawn separately for subsets of data using hue mapping
sns.displot(data=df, x='charges', hue='smoker')
```

As can be seen from the below graph, there is not much difference of distribution of data with respect to gender.

```python
# sometimes its difficult to visualize 2 histogram merged on one another
# hence, we can use multiple and element to show graph differently
sns.displot(data=df, x='charges', hue='sex', multiple='stack')
sns.displot(data=df, x='charges', hue='sex', element='step')
```

In the similar fashion, region also does not have any affects on the price of the insurance premium

```python
sns.displot(data=df, x='charges', hue='region', multiple='stack')
```

```python
# kde plot is also helpful if subplots are ovarlapping
sns.displot(data=df, x='charges', kind='kde', hue='sex')
sns.displot(data=df, x='charges', kind='kde', hue='region')
```

```python
# multipl subplots can be plotted side by side using col attribute
sns.displot(data=df, x='charges', hue='smoker', col='sex')
```

#### Does bmi affects insurance charges differently for male and female?

```python
sns.displot(data=df, x='bmi', y='charges', col='sex')
```

```python
sns.scatterplot(data=df, x='bmi', y='charges', hue='sex')
```

```python
sns.regplot(data=df[df['sex'] == 'male'], x='bmi', y='charges')
sns.regplot(data=df[df['sex'] == 'female'], x='bmi', y='charges', color='g')
```

#### Does premium charges increases with age?

```python
sns.regplot(data=df[df['smoker'] == 'yes'], x='age', y='charges', color='g')
sns.regplot(data=df[df['smoker'] == 'no'], x='age', y='charges', color='b')
```
