```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
%matplotlib inline
```

```python
os.chdir('C://Users//panka//github_repo//datasets//kaggle_dataset/')
os.getcwd()
df = pd.read_csv('insurance.csv')
df.head()
```

```python
# sns distplot has been deprecated, using displot
# plots can be drawn separately for subsets of data using hue mapping
sns.displot(data=df, x='charges', hue='smoker')
```

```python
# sometimes its difficult to visualize 2 histogram merged on one another
# hence, we can use multiple and element to show graph differently
sns.displot(data=df, x='charges', hue='sex', multiple='stack')
sns.displot(data=df, x='charges', hue='sex', element='step')
```

```python
sns.displot(data=df, x='charges', hue='region', multiple='stack')
```

```python
# kde plot is also helpful if subplots are ovarlapping
sns.displot(data=df, x='charges', kind='kde', hue='sex')
sns.displot(data=df, x='charges', kind='kde', hue='region')
```

```python
# multipl subplots can be plotted side by side using col attribute
sns.displot(data=df, x='charges', hue='smoker', col='sex')
```
