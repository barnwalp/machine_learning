```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
os.chdir('C:/Users/panka/github_repo/datasets/kaggle_dataset/')
os.getcwd()
df = pd.read_csv('rainfall in india 1901-2015.csv')
df.head()
```

```python
df['SUBDIVISION'].value_counts()
```

```python
# Using matplotlib to configure the seaborn plots
plt.figure(figsize=(40, 22))
plt.subplot(211)
plt.title('Rainfall in Jharkhand (1901-2015)')
plt.xticks(rotation=90)
sns.barplot(data=df[df['SUBDIVISION'] == 'JHARKHAND'], x='YEAR', y='ANNUAL')

plt.subplot(212)
plt.title('Rainfall in Bihar (1901-2015)')
plt.xticks(rotation=90)
sns.barplot(data=df[df['SUBDIVISION'] == 'BIHAR'], x='YEAR', y='ANNUAL')
```

```python
plt.figure(figsize=(20, 8))
plt.title('Historical rainfall in Jharkhand (1901-2015)')

# seaborn's lineplot returns an axis-object(not a figure)
ax = sns.lineplot(data=df[df['SUBDIVISION'] == 'JHARKHAND']
                )
ax.set(xlabel='index', ylabel='year')
```

```python
# Note: How labels and legend work in matplotlib
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x), '-b', label='Sine')
ax.plot(x, np.cos(x), '--r', label='Cosine')
#ax.axis('equal')
ax.legend(loc='upper right', frameon=False)
plt.xlabel('x')
plt.ylabel('y')
```
